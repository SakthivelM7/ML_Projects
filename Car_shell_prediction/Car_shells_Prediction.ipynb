# c covert object datatype to integer type data type

```python
df['price'] = pd.to_numeric(df['price'],errors='coerce')
df['horsepower'] = pd.to_numeric(df['horsepower'],errors='coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'],errors='coerce')
df['bore'] = pd.to_numeric(df['bore'],errors='coerce')
df['stroke'] = pd.to_numeric(df['stroke'],errors='coerce')
df
```

# d convert string to integer

```python
x = df['engine-location'].unique()
eng_loc = {}
for i in range(len(x)):
    eng_loc[x[i]] = i+1
df['engine-location'].replace(eng_loc,inplace=True)
df['engine-location']=pd.to_numeric(df['engine-location'])
df
```

```python
cylinder = df['num-of-cylinders'].unique()
noc = {}
for i in range(len(cylinder)):
    noc[cylinder[i]] = i+1
df['num-of-cylinders'].replace(noc,inplace=True)
df['num-of-cylinders']=pd.to_numeric(df['num-of-cylinders'])
df
```

```python
car = df['make'].unique()
make_car = {}
for i in range(len(car)):
    make_car[car[i]] = i+1
df['make'].replace(make_car,inplace=True)
df['make']=pd.to_numeric(df['make'])
df
```

```python
fuel= df['fuel-type'].unique()
y = {}
for i in range(len(fuel)):
    y[fuel[i]] = i+1
df['fuel-type'].replace(y,inplace=True)
df['fuel-type']=pd.to_numeric(df['fuel-type'])
df
```

```python
asp = df['aspiration'].unique()
z = {}
for i in range(len(asp)):
    z[asp[i]] = i+1
df['aspiration'].replace(z,inplace=True)
df['aspiration']=pd.to_numeric(df['aspiration'])
df
```

```python
doors= df['num-of-doors'].unique()
a = {}
for i in range(len(doors)):
    a[doors[i]] = i+1
df['num-of-doors'].replace(a,inplace=True)
df['num-of-doors']=pd.to_numeric(df['num-of-doors'])
df
```

```python
body = df['body-style'].unique()
a = {}
for i in range(len(body)):
    a[body[i]] = i+1
df['body-style'].replace(a,inplace=True)
df['body-style']=pd.to_numeric(df['body-style'])
df
```

```python
drive = df['drive-wheels'].unique()
a = {}
for i in range(len(drive)):
    a[drive[i]] = i+1
df['drive-wheels'].replace(a,inplace=True)
df['drive-wheels']=pd.to_numeric(df['drive-wheels'])
df
```

```python
engine = df['engine-type'].unique()
a = {}
for i in range(len(engine)):
    a[engine[i]] = i+1
df['engine-type'].replace(a,inplace=True)
df['engine-type']=pd.to_numeric(df['engine-type'])
df
```

```python
system = df['fuel-system'].unique()
a = {}
for i in range(len(system)):
    a[system[i]] = i+1
df['fuel-system'].replace(a,inplace=True)
df['fuel-system']=pd.to_numeric(df['fuel-system'])
df
```

```python
fill_dict = {'bore':df['bore'].mean(),
             'stroke':df['stroke'].mean(),
            'price':df['price'].mean()}
df.fillna(fill_dict,inplace=True)
```

# split the data for training

```python
from sklearn.model_selection import train_test_split
x=df.drop('price',axis=1)
y = df['price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
df1 = df.drop('price',axis=1)
col = df.columns
for i in col:
    count= len(df.loc[df[i]==0])
    print(f'{i} count is {count}')
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
linear_predict = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import math
lmodel_mse = mean_squared_error(linear_predict,y_test)
math.sqrt(lmodel_mse)
```

# train lasso algorithm

```python
from sklearn.linear_model import Lasso

lass_model = Lasso(alpha = 0.5)
lass_model.fit(x_train,y_train)

lasso_predict=lass_model.predict(x_test)
from sklearn.metrics import mean_squared_error

lasso_mse = mean_squared_error(lasso_predict,y_test)
math.sqrt(lasso_mse)
```

# training ridge algorithm

```python
from sklearn.linear_model import Ridge

rmodel = Ridge()
rmodel.fit(x_train,y_train)

ridge_predict=rmodel.predict(x_test)
from sklearn.metrics import mean_squared_error

ridge_mse = mean_squared_error(ridge_predict,y_test)
math.sqrt(ridge_mse)
```

# conclusion by comparing all algorithem ridge algorithem is giving best performance. so i will prefer ridge algorithem for predicting  the vehical cost
